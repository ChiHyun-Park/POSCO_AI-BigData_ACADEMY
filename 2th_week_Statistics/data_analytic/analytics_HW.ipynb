```python
# randn을 호출한다.
from numpy.random import randn
```

```python
# 샘플의 데이터를 1234중의 하나만 나오게 한다.
np.random.seed(1234)

# 랜덤 함수로 df데이터를 생성
df = pd.DataFrame(randn(10,4), columns=['A','B','C','D'])
df
```

```python
# slice 를 이용하여 (1) B열 선택 (2) B,D열 선택 (3) 4-8행 선택한다.
# (1) B열 선택
df[['B']]
```

```python
# (2) B,D열 선택
df[['B','D']]
```

```python
# (3) 4-8행 선택
df[4:9]
```

```python
# loc, iloc 을 이용하여 데이터 선택한다.
# (1) loc으로 2-8행, A-C열 선택
df.loc[2:8,['A','C']]
```

```python
# (2) loc으로 A-C열 선택
df.loc[:,['A','C']]
```

```python
# (3) loc으로 3행 선택
df.loc[3:3]
```

```python
# (1) iloc으로 0-6행, A열 선택
df.iloc[0:7,[0]]
```

```python
# (2) iloc으로 A열 선택
df.iloc[:,[0]]
```

```python
# (3) iloc으로 2-6행 선택
df.iloc[2:7]
```

```python
# (4) iloc으로 5행 선택
df.iloc[5:6]
```

```python
# query를 이용하여 데이터를 조작한다.
# (1) D>0.5 데이터 선택
df.query('D > 0.5')
```

```python
# (2) B>D 데이터 선택(slice 표현으로 선택)
df.query('B>D')
```

```python
# (3) A>0.5 그리고 D<0.0 데이터 선택
df.query('A>0.5 & D<0.0')
```

```python
# (4) A>0.5 혹은 D<0.0 데이터 선택
df.query('A>0.5 | D<0.0')
```

## 데이터 변환

```python
# 결론
# groupby 와 agg 를 사용하여 stacked table을 만들었지만
# 이를 unstacking한다음 pivot함수를 써서 결과값을 구하는법을 찾지못했다.
# 다만 unstacking의 파라미터를 조정하여 원하는 결과값을 얻을 수 있었다.
```

```python
df = pd.read_csv('C:/Users/ParkChihyun/Desktop/2주차 교육자료/데이터 Handling/FITNESS.csv',encoding="utf-8")
df
```

```python
# 분석에 필요한 컬럼들만 뽑아오자
df2 = df.loc[:,['AGEGROUP','GENDER','WEIGHT']]
df2
```

```python
# https://uincity.tistory.com/256 (groupby와 평균 공식설명)
# https://jamesrledoux.com/code/group-by-aggregate-pandas (multiple grouping이 가능함)
# https://pandas.pydata.org/docs/user_guide/reshaping.html (판다스홈페이지 pivot 설명)
# 나이대와 성별에 따른 몸무게의 평균을 구한다. (groupby와 mean을 사용)
df3 = df.groupby(["AGEGROUP","GENDER"]).agg({"WEIGHT":[np.mean]})
# 새롭게 구한 평균 몸무게의 컬럼 명을 지정해준다.
df3.columns = ['WEIGHT_MEAN']
df3
# stacked 된 상태라서 unstacked해준다
# unstack과정에서 상위레벨인덱스가 인덱스로 유지, 하위레벨인덱스가 컬럼이 된다.
unstacked_df3 = df3.unstack()
unstacked_df3
```

```python
unstacked_df4 = df3.unstack(level=1)
unstacked_df4
```

```python
# pop_seoul=pop_seoul.astype({"인구수":"int","한국인":"int","외국인":"int","고령자":"int"})
# data type을 바꿔주었다.
pop_seoul['인구수']=pop_seoul.인구수.str.replace(',','').astype('int64')
pop_seoul['한국인']=pop_seoul.한국인.str.replace(',','').astype('int64')
pop_seoul['외국인']=pop_seoul.외국인.str.replace(',','').astype('int64')
pop_seoul['고령자']=pop_seoul.고령자.str.replace(',','').astype('int64')
```

```python
pop_seoul['외국인'] = pop_seoul['외국인'].map('{:,}'.format)
```
