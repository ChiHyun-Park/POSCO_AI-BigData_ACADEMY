```python
# 학생별 각각의 점수로 이루어진 이중리스트에 평균과 학점을 추가하자.
def sorted_list():
    for student in range(len(final_list)):
        avg = ((float(final_list[student][2])+float(final_list[student][3]))/2)  # float으로 해야 소수점 첫째자리까지 나옴.
        final_list[student].append(avg)
        
        if avg > 90:
            grade = 'A'
            final_list[student].append(grade)
        elif 80<=avg<90:
            grade = 'B'
            final_list[student].append(grade)
        elif 70<=avg<80:
            grade = 'C'
            final_list[student].append(grade)
        elif 60<=avg<70:
            grade = 'D'
            final_list[student].append(grade)
        else:
            grade = 'F'
            final_list[student].append(grade)
            
     
        student_list.append(final_list[student][1]) # 빈리스트에 학생이름
        number_list.append(final_list[student][0]) # 빈리스트에 학생학번 하나씩 넣어준다. 근데 이거 앞에서 빈 리스트로 정의했으니깐 전역변수라서 다시 나오겠지?

#     for student in range(len(final_list)):
#         student_list.append(final_list[student][1]) # 빈리스트에 학생이름
#         number_list.append(final_list[student][0]) # 빈리스트에 학생학번 하나씩 넣어준다. 근데 이거 앞에서 빈 리스트로 정의했으니깐 전역변수라서 다시 나오겠지?
    
       
        
# 이 과정을 통해 final_list 에 평균과 grade가 추가되었다. 
# stduent_list도 업데이트해주자 
```

```python
# show누르면 차트 최종본 나오도록!! -> sorted_list() 거치면 아래에 어떻게 출력해야할까
def show_chart():   # 호출되기 전에 이미 위에서 sorted_list() 거쳤으니깐 평균 Grade 추가된 상태
    print("Student \t Name \t     Midterm   Final  Average  Grade")
    print("---------------------------------------------------------------------")
    final_list.sort(key=lambda e : e[4], reverse=True)  # final_list 를 평균에 따라 내림차순 정렬
    for i in range(len(final_list)):
        finalindex = len(final_list[i])-1
        for x in range(len(final_list[i])-1):
            print(final_list[i][x], end='\t')
        print(final_list[i][5], end='\n')
```

```python
def remove_grade():
    number_list = list()
    final_list.sort(key=lambda e : e[4], reverse=True)
    for student in range(len(final_list)):
        number_list.append(final_list[student][0])
    for i in range(len(final_list)):
        del[final_list[i][4]]
        del[final_list[i][4]]
    new_command  = input("Student ID: ")
    if number_list == []:
        print("List is empty.")
    else:
        if new_command not in number_list:
            print("NO SUCH PERSON.")
        else:
            delete_number = number_list.index(new_command)
            del[final_list[delete_number]]

            sorted_list()
            print("Student removed.")
        
```

```python
def search_grade():
    count=0
    final_list.sort(key=lambda e : e[4], reverse=True)  # 평균에 따라 정렬된 상태!!
    grade_index = list()    # 찾고자하는 grade를 가진 학생의 단일리스트 모음. 즉, 이중리스트!
    grade_kind = ['A','B','C','D','F']   #학점의 종류
    new_command  = input("Grade to search: ")
    if new_command in grade_kind:
        for i in range(len(final_list)):
            if new_command == final_list[i][5]:
                count+=1
                grade_index.append(final_list[i])   
        if count == 0:
            print("NO RESULTS")
        else:
            print("Student \t Name \t     Midterm   Final  Average  Grade")
            print("---------------------------------------------------------------------")
            for i in range(len(grade_index)): 
                for x in range(len(grade_index[i])-1):  
                    print(grade_index[i][x],end = '\t')
                print(grade_index[i][5],end ='\n') 
    else:
        pass
```

```python
def add_student():
    new_command  = input("Student ID: ")
    if new_command in number_list:
        print("ALREADY EXISTS.")
    else:
        new_new_command  = input("Name: ")
        new_new_new_command  = int(input("Midterm Score: "))
        new_new_new_new_command  = int(input("Final Score: "))
        # 새로운 학생의 빈리스트
        new_student = list()                               
        new_student = [new_command, new_new_command, new_new_new_command, new_new_new_new_command]
        # 기존의 final_list의 평균과 학점 제거
        for i in range(len(number_list)):  #5명이면 0부터 4까지~
            del final_list[i][4]
            del final_list[i][4]
        final_list.append(new_student)
        sorted_list()
        print("Student added.")
```

```python
def change_score():
    new_command  = input("Student ID: ")
    if new_command not in number_list:
        print("NO SUCH PERSON")
        
    else:
        target_index = int(number_list.index(new_command))  # 학번있을경우 몇번째 사람인지 index알려줌.
        new_new_command  = input("Mid/Final? ")
        
        # 중간 입력했을 때...
        if new_new_command.lower() == "mid":
            new_new_new_command  = int(input("Input new score: "))
            if new_new_new_command < 0 or new_new_new_command > 100:
                print("No such score")
                
            else:
                print("Student \t Name \t     Midterm   Final  Average  Grade")
                print("---------------------------------------------------------------------")
                for x in range(len(final_list[target_index])-1):   # index  0부터 4까지~~
                    print(final_list[target_index][x],end = '\t')
                print(final_list[target_index][5],end ='\n')
                print("Score changed.")
                final_list[target_index][2] = new_new_new_command
                del final_list[target_index][4]
                del final_list[target_index][4]   # 평균과 grade를 지운 상태에서 sorted_list해야 새로운 평균 학점이 생긴다. 또한 바로위에 4번인덱스지우면 5번인덱스가 4번인덱스로 내려와서 이렇게 표현한다. 
                sorted_list()
                for x in range(len(final_list[target_index])-1):   # index  0부터 4까지~~
                    print(final_list[target_index][x],end = '\t')
                print(final_list[target_index][5],end ='\n')
                
        # 기말 입력했을 때...
        elif new_new_command.lower() == "final":
            new_new_new_command  = int(input("Input new score: "))
            if new_new_new_command < 0 or new_new_new_command > 100:
                print("No such score")
            else:
                print("Student \t Name \t     Midterm   Final  Average  Grade")
                print("---------------------------------------------------------------------")
                for x in range(len(final_list[target_index])-1):   # index  0부터 4까지~~
                    print(final_list[target_index][x],end = '\t')
                print(final_list[target_index][5],end ='\n')
                print("Score changed.")
                final_list[target_index][2] = new_new_new_command
                del final_list[target_index][4]
                del final_list[target_index][4]   # 평균과 grade를 지운 상태에서 sorted_list해야 새로운 평균 학점이 생긴다. 또한 바로위에 4번인덱스지우면 5번인덱스가 4번인덱스로 내려와서 이렇게 표현한다. 
                sorted_list()
                for x in range(len(final_list[target_index])-1):   # index  0부터 4까지~~
                    print(final_list[target_index][x],end = '\t')
                print(final_list[target_index][5],end ='\n')
       
                
        else:
            pass
        
```

```python
def search_name():
    new_command  = input("Student ID: ")
    if new_command not in number_list:
        print("NO SUCH PERSON")
    else:   # 학번 list에 학번과 일치한다면~
        print("Student \t Name \t     Midterm   Final  Average  Grade")
        print("---------------------------------------------------------------------")
        target_index = int(number_list.index(new_command))
        for i in final_list[target_index]:
            print(i, end='\t')   # list성분 하나씩 나오니깐 i로 표깅
```

```python
# main함수 ~~~ 나중에 main함수 추가해서 잘되나 테스트해보자. 

import os
input_file = input("Enter a file name : ")

final_list = list()     # 학생 종합정보 모음 이중리스트
student_list = list()   # 학생 이름 모음 단일리스트
number_list = list()    # 학생 학번 모음 단일리스트

if (os.path.exists(input_file)):
    fr = open(input_file, 'r')
    print('load success', '\n')
    
    for line in fr:
        new_line = line.replace("\n",'')
        final_list.append(new_line.split("\t")) # 성과 이름사이의 공간은 \t으로 인식 안해서 고려할필요 없음.
        
elif input_file == '':
    input_file = 'students.txt'
    fr = open(input_file, 'r')
    print('access to students.txt','n')
    

sorted_list()   


while True:
    command = input("# ")             # #한칸뛰고 입력받는 공간 생김
    if command.lower() == "show":    # 함수 정의할 때 ()비워두면 나중에 함수 호출할때도 ()만 써도 호출이 된다. 
        show_chart()
    elif command.lower() == 'search':
        search_name()
    elif command.lower() == 'changescore':
        change_score()
    elif command.lower() == 'add':
        add_student()
    elif command.lower() == 'searchgrade':
        search_grade()
    elif command.lower() == 'remove':
        remove_grade()
    elif command.lower() == "quit":
        quit_ = input('Save? [yes/no] ')
        if quit_ == 'yes':
            file_name = input('File name: ')
            f = open(file_name, 'w')
            final_list.sort(key=lambda e : e[4], reverse=True)
            for person in range(len(final_list)):
                data = '%s\t%s\t%s\t%s\n' %(final_list[person][0],final_list[person][1],final_list[person][2],final_list[person][3])
                f.write(data)
            f.close()
            break
        elif quit_ == 'no':
            print("seeya")
            break
        else:
            continue
    else:
        continue
```
